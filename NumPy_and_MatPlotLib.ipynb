# Curve-2 : y = 5 - 3x

Plot `y = 5 - 3x` for x âˆˆ [-5,5]. Sample 20 equally spaced points in this interval, inclusive of endpoints. Add a title to the plot which has the equation of curve.

```python
x = np.linspace(-5,5,20)
y = 5 - 3*x
plt.scatter(x,y);
plt.plot(x,y);
plt.title("y = 5 - 3x");
```

# Curve-3 : $y = x^{2}$

Plot $y = x^{2}$ for x âˆˆ [-1,1].

*   5 equally spaced points in the interval, inclusive of endpoints
*   10 equally spaced points in the interval, inclusive of endpoints
*   20 equally spaced points in the interval, inclusive of endpoints
*   50 equally spaced points in the interval, inclusive of endpoints

Add title and axis labels to the plot





```python
x = np.linspace(-1,1,5)
y = x**2
plt.scatter(x,y);
plt.plot(x,y);
plt.title("$y = x^{2}$");
plt.xlabel("x");
plt.ylabel("y");
```

```python
x = np.linspace(-1,1,10)
y = x**2
plt.scatter(x,y);
plt.plot(x,y);
plt.title("$y = x^{2}$");
plt.xlabel("x");
plt.ylabel("y");
```

```python
x = np.linspace(-1,1,20)
y = x**2
plt.scatter(x,y);
plt.plot(x,y);
plt.title("$y = x^{2}$");
plt.xlabel("x");
plt.ylabel("y");
```

```python
x = np.linspace(-1,1,50)
y = x**2
plt.plot(x,y);
plt.title("$y = x^{2}$");
plt.xlabel("x");
plt.ylabel("y");
plt.axhline(color = 'black', linestyle = '--' ,linewidth = 0.8);
plt.axvline(color = 'black', linestyle = '--' ,linewidth = 0.8);
```

***Note that as you increase the number of equally spaced points, the graph gets smoother. The shape of parabola gets more defined***

# Curve-4 : Exercise

Plot `y = 3` and `x = 5` on the same plot. Color the first one blue and second one red. Limit the plot to the region x âˆˆ [0,6] and y âˆˆ [0,6]

# Subplots

`plt.subplot(r,c,ind)` divides the plotting region into `r x c` individual plotting windows and selects the plot at index `ind`. The index is set to 1 for the top-left plot and increases by one point from left to right and top to bottom. The top-left plot window has index 1 and the last window has index c.

$y = ln x$, $x âˆˆ (\frac{1}{e^{2}}$, $e^{3}$)

$y = e^{x}$, $x âˆˆ (0, 3$)

$y = sin x, x âˆˆ [0,2Ï€]$

$y = \frac{1}{x}, x \in (1,5)$


```python
plt.rcParams['figure.figsize'] = [8,8]
```

```python
plt.subplot(2,2,1);
x = np.linspace(1/np.e**2,np.e**3)
y = np.log(x)
plt.plot(x,y);
plt.title("$y = ln x$");

plt.subplot(2,2,2);
x = np.linspace(0,3)
y = np.exp(x)
plt.plot(x,y);
plt.title("$y = e^{x}$");

plt.subplot(2,2,3);
x = np.linspace(0,2*np.pi)
y = np.sin(x)
plt.plot(x,y);
plt.title("$y = sin x$");

plt.subplot(2,2,4)
x = np.linspace(1,5)
y = 1/x
plt.plot(x,y);
plt.title(r'$y = 1/x$');
```
